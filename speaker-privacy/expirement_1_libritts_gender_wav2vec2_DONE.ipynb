```python
################################################################################
# 1. INSTALL REQUIRED LIBRARIES (if not already installed)
################################################################################
# !pip install datasets
# !pip install transformers
# !pip install jiwer
# !pip install torchaudio
# !pip install scikit-learn
# !pip install matplotlib

################################################################################
# 2. IMPORTS
################################################################################
import torch
import torchaudio
import numpy as np
import random
import matplotlib.pyplot as plt

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from transformers import (
    Wav2Vec2ForCTC,
    Wav2Vec2Processor,
    TrainingArguments,
    Trainer
)
from dataclasses import dataclass
from typing import Any, Dict, List, Union
from collections import defaultdict

################################################################################
# 3. LOAD THE DATASET
################################################################################
dataset = load_dataset("HamdanXI/libritts_dev_dataset_gender")["train"]
# dataset has columns: ["audio", "gender", "speaker_id", "text_original", ...]

print("Full dataset size:", len(dataset))
print("Columns:", dataset.column_names)
print("Unique speaker_ids:", len(set(dataset["speaker_id"])))

################################################################################
# 4. CHOOSE 20 SPEAKERS (10 FEMALE + 10 MALE)
################################################################################
male_speakers = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "M"})
female_speakers = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "F"})

random.seed(42)
chosen_male = random.sample(male_speakers, 10)
chosen_female = random.sample(female_speakers, 10)

chosen_speakers = chosen_male + chosen_female
print(f"Chosen 10 male speaker_ids: {chosen_male}")
print(f"Chosen 10 female speaker_ids: {chosen_female}")

# Filter dataset to keep only those 20 speakers
filtered_ds = dataset.filter(lambda x: x["speaker_id"] in chosen_speakers)

################################################################################
# 5. LOAD PROCESSOR
################################################################################
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")

################################################################################
# 6. PREPARE BATCH FUNCTION
################################################################################
def prepare_batch(batch):
    """
    1) Resample audio to 16k if needed.
    2) Extract raw (unpadded) input_values via processor.feature_extractor(...).
    3) Tokenize text via processor.tokenizer(...).
    """
    # 1) Resample audio
    sr = batch["audio"]["sampling_rate"]
    audio_array = batch["audio"]["array"]
    wave_tensor = torch.tensor(audio_array, dtype=torch.float32)
    if sr != 16000:
        wave_tensor = torchaudio.functional.resample(wave_tensor, sr, 16000)
    wave_16k = wave_tensor.numpy()

    # 2) Audio feature extraction (no return_tensors here; we rely on the data collator)
    audio_features = processor.feature_extractor(
        wave_16k,
        sampling_rate=16000,
    )
    # audio_features["input_values"] is a list of shape [1, audio_length]

    # 3) Text tokenization (no return_tensors here)
    text_tokens = processor.tokenizer(
        batch["text_original"]
    )
    # text_tokens["input_ids"] is a list of token IDs (no padding yet)

    return {
        "speaker_id":    batch["speaker_id"],
        "gender":        batch["gender"],
        "input_values":  audio_features["input_values"][0],
        "labels":        text_tokens["input_ids"]
    }

# Keep only needed columns
keep_cols = ["speaker_id", "gender", "input_values", "labels"]
filtered_ds = filtered_ds.map(
    prepare_batch,
    remove_columns=[col for col in filtered_ds.column_names if col not in keep_cols]
)

################################################################################
# 7. DATA COLLATOR
################################################################################
@dataclass
class DataCollatorCTCWithPadding:
    """
    1) Pads audio via processor.feature_extractor.pad(...)
    2) Pads labels via processor.tokenizer.pad(...)
    3) Converts pad_token_id -> -100 for CTC
    """
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True

    def __call__(self, features: List[Dict[str, Any]]) -> Dict[str, torch.Tensor]:
        # Separate out inputs/labels
        input_values = [f["input_values"] for f in features]
        labels = [f["labels"] for f in features]

        # 1) Pad audio
        audio_batch = self.processor.feature_extractor.pad(
            {"input_values": input_values},
            padding=self.padding,
            return_tensors="pt",
        )

        # 2) Pad labels
        labels_batch = self.processor.tokenizer.pad(
            {"input_ids": labels},
            padding=self.padding,
            return_tensors="pt",
        )

        # Convert pad_token_id -> -100
        labels_tensor = labels_batch["input_ids"].masked_fill(
            labels_batch["input_ids"] == self.processor.tokenizer.pad_token_id, -100
        )

        return {
            "input_values": audio_batch["input_values"],
            "labels":       labels_tensor
        }

data_collator = DataCollatorCTCWithPadding(processor=processor, padding=True)

################################################################################
# 8. TRAIN ON SINGLE SAMPLE
################################################################################
def train_on_single_sample(
    sample,
    base_model_name="facebook/wav2vec2-base-960h",
    num_train_epochs=1,
    device="cuda" if torch.cuda.is_available() else "cpu"
):
    """
    Fine-tune Wav2Vec2ForCTC on a single sample:
      "input_values" and "labels"
    Returns the fine-tuned model on CPU (moved back after training).
    """
    from datasets import Dataset

    # Make a tiny dataset with 1 row
    single_ds = Dataset.from_dict({
        "input_values": [sample["input_values"]],
        "labels":       [sample["labels"]]
    })

    # Load a fresh base model
    model = Wav2Vec2ForCTC.from_pretrained(base_model_name)
    # Send to device (GPU if available)
    model.to(device)
    # Optionally freeze feature encoder
    model.freeze_feature_encoder()

    training_args = TrainingArguments(
        output_dir="temp_model",
        per_device_train_batch_size=1,
        num_train_epochs=num_train_epochs,
        evaluation_strategy="no",
        logging_steps=5,
        learning_rate=1e-5,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=single_ds,
        tokenizer=processor.feature_extractor,  # optional for single-sample
    )

    # Train
    trainer.train()

    # Move model back to CPU to ensure no device mismatch
    model.to("cpu")
    return model

################################################################################
# 9. EXTRACT LAYERWISE DIFFERENCES (MEAN & STD)
################################################################################
def compute_layerwise_diff(new_model, base_model, num_layers=12):
    """
    Compare new_model vs. base_model for each Wav2Vec2 encoder layer.
    Returns two lists (of length num_layers):
      mean_diffs[layer_index], std_diffs[layer_index]

    Both models assumed on the same device (CPU).
    """
    layerwise_params = defaultdict(list)

    base_dict = dict(base_model.named_parameters())
    new_dict  = dict(new_model.named_parameters())

    for name, base_param in base_dict.items():
        if ("wav2vec2.encoder.layers" in name) and (name in new_dict):
            new_param = new_dict[name]
            parts = name.split(".")
            # e.g. ["wav2vec2","encoder","layers","3","attention","k_proj","weight"]
            try:
                layer_idx = int(parts[3])  # parse "3"
            except ValueError:
                continue

            if layer_idx < num_layers:
                # Both base_param & new_param should be on CPU now
                diff = (new_param - base_param).detach().cpu().view(-1)
                layerwise_params[layer_idx].append(diff)

    mean_diffs = []
    std_diffs  = []
    for i in range(num_layers):
        if i in layerwise_params:
            diffs_cat = torch.cat(layerwise_params[i], dim=0)
            mean_diffs.append(diffs_cat.abs().mean().item())
            std_diffs.append(diffs_cat.abs().std().item())
        else:
            mean_diffs.append(0.0)
            std_diffs.append(0.0)

    return mean_diffs, std_diffs

################################################################################
# 10. MAIN LOOP: PICK 1 SAMPLE PER SPEAKER, TRAIN, COMPARE
################################################################################
# Convert filtered dataset to list
all_samples = list(filtered_ds)

# Group samples by speaker_id
samples_by_speaker = defaultdict(list)
for ex in all_samples:
    samples_by_speaker[ex["speaker_id"]].append(ex)

# For each of our 20 chosen speakers, pick a single row
speaker_to_sample = {}
for spk in chosen_speakers:
    speaker_rows = samples_by_speaker[spk]
    random.seed(1234)
    chosen_row = random.choice(speaker_rows)
    speaker_to_sample[spk] = chosen_row

# Load the *BASE* model once for reference (on CPU)
base_model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h").eval()
base_model.to("cpu")

results = {}

# By default, wav2vec2-base has 12 layers (0..11).
# We only want 11 layers => skip layer 0 => keep [1..11].
LAYER_OFFSET = 1
NUM_LAYERS   = 12
LAYERS_TO_USE = list(range(LAYER_OFFSET, NUM_LAYERS))  # [1..11]

for spk_id in chosen_speakers:
    print(f"\n=== Fine-tuning for speaker: {spk_id} ===")
    sample = speaker_to_sample[spk_id]

    # Fine-tune on single sample (GPU if available, then move to CPU)
    trained_model = train_on_single_sample(sample)

    # Compute differences vs. base (both on CPU)
    mean_diffs, std_diffs = compute_layerwise_diff(trained_model, base_model, NUM_LAYERS)

    # Keep only layers [1..11]
    mean_diffs = [mean_diffs[i] for i in LAYERS_TO_USE]
    std_diffs  = [std_diffs[i] for i in LAYERS_TO_USE]

    results[spk_id] = {
        "gender": sample["gender"],
        "mean_diffs": mean_diffs,
        "std_diffs": std_diffs
    }

    del trained_model
    torch.cuda.empty_cache()

################################################################################
# 11. PLOTTING
################################################################################
# 3 plots:
#   1) Mean difference (20 lines)
#   2) Std difference  (20 lines)
#   3) Mean+Std difference (20 lines)
#   M = blue, F = pink

layer_indices = list(range(LAYER_OFFSET, NUM_LAYERS))  # [1..11]

fig1, ax1 = plt.subplots(figsize=(8,5))
fig2, ax2 = plt.subplots(figsize=(8,5))
fig3, ax3 = plt.subplots(figsize=(8,5))

for spk_id, data in results.items():
    gender = data["gender"]
    color = "blue" if gender == "M" else "pink"

    mean_vals = data["mean_diffs"]
    std_vals  = data["std_diffs"]
    sum_vals  = [m + s for m, s in zip(mean_vals, std_vals)]

    # Plot 1: mean
    ax1.plot(layer_indices, mean_vals, color=color, alpha=0.7, label=gender)

    # Plot 2: std
    ax2.plot(layer_indices, std_vals, color=color, alpha=0.7, label=gender)

    # Plot 3: mean+std
    ax3.plot(layer_indices, sum_vals, color=color, alpha=0.7, label=gender)

# Deduplicate legend entries
def deduplicate_legend(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique_labels = []
    unique_handles = []
    for h, l in zip(handles, labels):
        if l not in unique_labels:
            unique_labels.append(l)
            unique_handles.append(h)
    ax.legend(unique_handles, unique_labels)

deduplicate_legend(ax1)
ax1.set_title("Mean Parameter Difference (Layers 1..11)")
ax1.set_xlabel("Layer Index")
ax1.set_ylabel("Mean Abs Diff")

deduplicate_legend(ax2)
ax2.set_title("Std Parameter Difference (Layers 1..11)")
ax2.set_xlabel("Layer Index")
ax2.set_ylabel("Std of Abs Diff")

deduplicate_legend(ax3)
ax3.set_title("Mean + Std Parameter Difference (Layers 1..11)")
ax3.set_xlabel("Layer Index")
ax3.set_ylabel("Mean + Std of Abs Diff")

plt.tight_layout()
plt.show()

################################################################################
# DONE
################################################################################
```

```python
################################################################################
# 1. INSTALL REQUIRED LIBRARIES (if not already installed)
################################################################################
# !pip install datasets
# !pip install transformers
# !pip install jiwer
# !pip install torchaudio
# !pip install scikit-learn
# !pip install matplotlib

################################################################################
# 2. IMPORTS
################################################################################
import torch
import torchaudio
import numpy as np
import random
import matplotlib.pyplot as plt

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from transformers import (
    Wav2Vec2ForCTC,
    Wav2Vec2Processor,
    TrainingArguments,
    Trainer
)
from dataclasses import dataclass
from typing import Any, Dict, List, Union
from collections import defaultdict, Counter

################################################################################
# 3. LOAD THE DATASET
################################################################################
dataset = load_dataset("HamdanXI/libritts_dev_dataset_gender")["train"]
# dataset columns: ["audio", "gender", "speaker_id", "text_original", ...]

print("Full dataset size:", len(dataset))
print("Columns:", dataset.column_names)
print("Unique speaker_ids:", len(set(dataset["speaker_id"])))

################################################################################
# 4. CHOOSE 20 "TRAIN" + 10 "TEST" SPEAKERS
################################################################################
# We have male/female lists:
male_speakers_all = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "M"})
female_speakers_all = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "F"})

random.seed(42)
# 1) pick 20 training: 10M + 10F
chosen_male_train = random.sample(male_speakers_all, 10)
chosen_female_train = random.sample(female_speakers_all, 10)

# Remove them from the big list so we don't re-pick them for test
for spk in chosen_male_train:
    male_speakers_all.remove(spk)
for spk in chosen_female_train:
    female_speakers_all.remove(spk)

# 2) pick 10 test: 5M + 5F
chosen_male_test = random.sample(male_speakers_all, 5)
chosen_female_test = random.sample(female_speakers_all, 5)

chosen_speakers_train = chosen_male_train + chosen_female_train
chosen_speakers_test  = chosen_male_test + chosen_female_test

print(f"Chosen training male:   {chosen_male_train}")
print(f"Chosen training female: {chosen_female_train}")
print(f"Chosen test male:       {chosen_male_test}")
print(f"Chosen test female:     {chosen_female_test}")

# Filter the dataset into two subsets: train_speakers, test_speakers
train_ds = dataset.filter(lambda x: x["speaker_id"] in chosen_speakers_train)
test_ds  = dataset.filter(lambda x: x["speaker_id"] in chosen_speakers_test)

################################################################################
# 5. LOAD PROCESSOR
################################################################################
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")

################################################################################
# 6. PREPARE BATCH FUNCTION
################################################################################
def prepare_batch(batch):
    """
    1) Resample audio to 16k if needed.
    2) Extract raw (unpadded) input_values via processor.feature_extractor(...).
    3) Tokenize text via processor.tokenizer(...).
    """
    sr = batch["audio"]["sampling_rate"]
    audio_array = batch["audio"]["array"]
    wave_tensor = torch.tensor(audio_array, dtype=torch.float32)
    if sr != 16000:
        wave_tensor = torchaudio.functional.resample(wave_tensor, sr, 16000)
    wave_16k = wave_tensor.numpy()

    # Audio feature extraction
    audio_features = processor.feature_extractor(wave_16k, sampling_rate=16000)
    # Text tokenization
    text_tokens = processor.tokenizer(batch["text_original"])

    return {
        "speaker_id":   batch["speaker_id"],
        "gender":       batch["gender"],
        "input_values": audio_features["input_values"][0],
        "labels":       text_tokens["input_ids"]
    }

keep_cols = ["speaker_id", "gender", "input_values", "labels"]
train_ds = train_ds.map(
    prepare_batch,
    remove_columns=[col for col in train_ds.column_names if col not in keep_cols]
)

test_ds = test_ds.map(
    prepare_batch,
    remove_columns=[col for col in test_ds.column_names if col not in keep_cols]
)

################################################################################
# 7. DATA COLLATOR
################################################################################
@dataclass
class DataCollatorCTCWithPadding:
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True

    def __call__(self, features: List[Dict[str, Any]]) -> Dict[str, torch.Tensor]:
        input_values = [f["input_values"] for f in features]
        labels       = [f["labels"]       for f in features]

        audio_batch = self.processor.feature_extractor.pad(
            {"input_values": input_values},
            padding=self.padding,
            return_tensors="pt",
        )

        labels_batch = self.processor.tokenizer.pad(
            {"input_ids": labels},
            padding=self.padding,
            return_tensors="pt",
        )

        labels_tensor = labels_batch["input_ids"].masked_fill(
            labels_batch["input_ids"] == self.processor.tokenizer.pad_token_id, -100
        )

        return {
            "input_values": audio_batch["input_values"],
            "labels":       labels_tensor
        }

data_collator = DataCollatorCTCWithPadding(processor=processor, padding=True)

################################################################################
# 8. TRAIN ON SINGLE SAMPLE
################################################################################
def train_on_single_sample(
    sample,
    base_model_name="facebook/wav2vec2-base-960h",
    num_train_epochs=1,
    device="cuda" if torch.cuda.is_available() else "cpu"
):
    """
    Fine-tune on a single sample (dictionary with "input_values", "labels").
    Returns the fine-tuned model on CPU.
    """
    from datasets import Dataset

    single_ds = Dataset.from_dict({
        "input_values": [sample["input_values"]],
        "labels":       [sample["labels"]]
    })

    model = Wav2Vec2ForCTC.from_pretrained(base_model_name)
    model.to(device)
    model.freeze_feature_encoder()

    training_args = TrainingArguments(
        output_dir="temp_model",
        per_device_train_batch_size=1,
        num_train_epochs=num_train_epochs,
        evaluation_strategy="no",
        logging_steps=5,
        learning_rate=1e-5,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=single_ds,
        tokenizer=processor.feature_extractor
    )

    trainer.train()
    model.to("cpu")
    return model

################################################################################
# 9. EXTRACT LAYERWISE DIFFERENCES
################################################################################
def compute_layerwise_diff(new_model, base_model, num_layers=12):
    """
    Compare new_model vs. base_model for each Wav2Vec2 encoder layer.
    Return two lists of length num_layers: mean_diffs, std_diffs
    (Both models on CPU).
    """
    from collections import defaultdict
    layerwise_params = defaultdict(list)

    base_dict = dict(base_model.named_parameters())
    new_dict  = dict(new_model.named_parameters())

    for name, base_param in base_dict.items():
        if ("wav2vec2.encoder.layers" in name) and (name in new_dict):
            new_param = new_dict[name]
            parts = name.split(".")
            try:
                layer_idx = int(parts[3])  # e.g. "3"
            except:
                continue

            if layer_idx < num_layers:
                diff = (new_param - base_param).detach().cpu().view(-1)
                layerwise_params[layer_idx].append(diff)

    mean_diffs, std_diffs = [], []
    for i in range(num_layers):
        if i in layerwise_params:
            diffs_cat = torch.cat(layerwise_params[i], dim=0)
            mean_diffs.append(diffs_cat.abs().mean().item())
            std_diffs.append(diffs_cat.abs().std().item())
        else:
            mean_diffs.append(0.0)
            std_diffs.append(0.0)

    return mean_diffs, std_diffs

################################################################################
# 10A. BUILD DIFFERENCE VECTORS FOR 20 TRAIN SPEAKERS
################################################################################
train_samples = list(train_ds)
samples_by_speaker_train = defaultdict(list)
for ex in train_samples:
    samples_by_speaker_train[ex["speaker_id"]].append(ex)

# We'll pick 1 row per speaker for the 20 training speakers
speaker_to_sample_train = {}
random.seed(1234)
for spk in chosen_speakers_train:
    spk_rows = samples_by_speaker_train[spk]
    chosen_row = random.choice(spk_rows)
    speaker_to_sample_train[spk] = chosen_row

# Load base model once
base_model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h").eval()
base_model.to("cpu")

LAYER_OFFSET = 1
NUM_LAYERS   = 12
LAYERS_TO_USE = list(range(LAYER_OFFSET, NUM_LAYERS))  # [1..11]

train_results = {}

print("\n=== Building difference vectors for 20 training speakers ===")
for spk_id in chosen_speakers_train:
    sample = speaker_to_sample_train[spk_id]
    print(f"Fine-tuning on speaker {spk_id} (gender={sample['gender']})")

    trained_model = train_on_single_sample(sample)
    mean_diffs, std_diffs = compute_layerwise_diff(trained_model, base_model, NUM_LAYERS)

    # Keep [1..11]
    mean_diffs = [mean_diffs[i] for i in LAYERS_TO_USE]
    std_diffs  = [std_diffs[i]  for i in LAYERS_TO_USE]

    train_results[spk_id] = {
        "gender": sample["gender"],
        "mean_diffs": mean_diffs,
        "std_diffs": std_diffs
    }

    del trained_model
    torch.cuda.empty_cache()

################################################################################
# 10B. AVERAGE MALE/ FEMALE DIFFERENCE VECTORS
################################################################################
male_vectors_mean  = []
male_vectors_std   = []
female_vectors_mean= []
female_vectors_std = []

for spk_id, data in train_results.items():
    if data["gender"] == "M":
        male_vectors_mean.append(data["mean_diffs"])
        male_vectors_std.append(data["std_diffs"])
    else:
        female_vectors_mean.append(data["mean_diffs"])
        female_vectors_std.append(data["std_diffs"])

# Convert to numpy for easy averaging
male_mean_avg   = np.mean(male_vectors_mean, axis=0)
male_std_avg    = np.mean(male_vectors_std,  axis=0)
female_mean_avg = np.mean(female_vectors_mean,axis=0)
female_std_avg  = np.mean(female_vectors_std, axis=0)

################################################################################
# 11. PLOT THE TRAINING SPEAKER DIFFERENCES
################################################################################
fig1, ax1 = plt.subplots(figsize=(6,4))
fig2, ax2 = plt.subplots(figsize=(6,4))
fig3, ax3 = plt.subplots(figsize=(6,4))

layer_indices = LAYER_OFFSET + np.arange(len(LAYERS_TO_USE))  # 1..11

for spk_id, data in train_results.items():
    gender = data["gender"]
    color = "blue" if gender == "M" else "pink"

    mean_vals = data["mean_diffs"]
    std_vals  = data["std_diffs"]
    sum_vals  = [m + s for m, s in zip(mean_vals, std_vals)]

    # Mean
    ax1.plot(layer_indices, mean_vals, color=color, alpha=0.7, label=gender)
    # Std
    ax2.plot(layer_indices, std_vals,  color=color, alpha=0.7, label=gender)
    # Mean+Std
    ax3.plot(layer_indices, sum_vals,  color=color, alpha=0.7, label=gender)

def deduplicate_legend(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique_labels = []
    unique_handles = []
    for h, l in zip(handles, labels):
        if l not in unique_labels:
            unique_labels.append(l)
            unique_handles.append(h)
    ax.legend(unique_handles, unique_labels)

deduplicate_legend(ax1)
ax1.set_title("Mean Difference (Training Speakers)")
ax1.set_xlabel("Layer Index")
ax1.set_ylabel("Mean Abs Diff")

deduplicate_legend(ax2)
ax2.set_title("Std Difference (Training Speakers)")
ax2.set_xlabel("Layer Index")
ax2.set_ylabel("Std of Abs Diff")

deduplicate_legend(ax3)
ax3.set_title("Mean+Std (Training Speakers)")
ax3.set_xlabel("Layer Index")
ax3.set_ylabel("Mean+Std of Abs Diff")

plt.tight_layout()
plt.show()

################################################################################
# 12. EVALUATE ON 10 TEST SPEAKERS
################################################################################
test_samples = list(test_ds)
samples_by_speaker_test = defaultdict(list)
for ex in test_samples:
    samples_by_speaker_test[ex["speaker_id"]].append(ex)

# We'll pick 1 row per speaker for the 10 test speakers
speaker_to_sample_test = {}
random.seed(1234)
for spk in chosen_speakers_test:
    spk_rows = samples_by_speaker_test[spk]
    chosen_row = random.choice(spk_rows)
    speaker_to_sample_test[spk] = chosen_row

# We'll define 2 classification methods:

def method_1_distance_to_avg(vec_mean, vec_std):
    """
    Compare new speaker's 'mean' difference vector to the average male/female
    difference vectors across all layers. Return "M" or "F".
    We use Euclidean distance in (mean_diffs) space:
        distance( new_mean, male_mean_avg )  vs. distance( new_mean, female_mean_avg )
    """
    # new_mean is the vector of length = 11
    new_mean = np.array(vec_mean)
    dist_male   = np.linalg.norm(new_mean - male_mean_avg)
    dist_female = np.linalg.norm(new_mean - female_mean_avg)
    return "M" if dist_male < dist_female else "F"

def method_2_single_layer_threshold(vec_mean, layer_idx=5):
    """
    Compare the difference at a single 'peak' layer (e.g. layer 5 if it stands out).
    We'll do a simple threshold between the average male vs. female at that layer.
    """
    # new_mean is the vector of length=11
    new_mean = np.array(vec_mean)
    avg_m = male_mean_avg[layer_idx]
    avg_f = female_mean_avg[layer_idx]

    # We'll pick a midpoint threshold:
    threshold = 0.5*(avg_m + avg_f)
    return "M" if new_mean[layer_idx] < threshold else "F"

# Evaluate
method_1_correct = 0
method_2_correct = 0
n_test = len(chosen_speakers_test)

print("\n=== Predicting on 10 Test Speakers ===")
for spk_id in chosen_speakers_test:
    sample = speaker_to_sample_test[spk_id]
    true_gender = sample["gender"]

    # Fine-tune single sample -> get difference vector
    test_model = train_on_single_sample(sample)
    mean_diffs, std_diffs = compute_layerwise_diff(test_model, base_model, NUM_LAYERS)
    mean_diffs = [mean_diffs[i] for i in LAYERS_TO_USE]
    # (We won't use std here in these simple classifiers, but you could.)

    # Method 1
    pred_1 = method_1_distance_to_avg(mean_diffs, std_diffs)
    if pred_1 == true_gender:
        method_1_correct += 1

    # Method 2 - let's pick a known "peak" layer, say layer 5 (which is index=4 in [1..11]).
    # If from the plots we suspect layer 6 is the biggest difference, that is index=5
    pred_2 = method_2_single_layer_threshold(mean_diffs, layer_idx=5)
    if pred_2 == true_gender:
        method_2_correct += 1

accuracy_1 = method_1_correct / n_test
accuracy_2 = method_2_correct / n_test
print(f"Method 1 (distance to avg) accuracy: {accuracy_1:.2%}")
print(f"Method 2 (single-layer threshold) accuracy: {accuracy_2:.2%}")

# We can also visualize the test speaker accuracies individually if we want
method_1_preds = {}
method_2_preds = {}
for spk_id in chosen_speakers_test:
    sample = speaker_to_sample_test[spk_id]
    true_gender = sample["gender"]

    test_model = train_on_single_sample(sample)
    mean_diffs, _ = compute_layerwise_diff(test_model, base_model, NUM_LAYERS)
    mean_diffs = [mean_diffs[i] for i in LAYERS_TO_USE]

    pred_1 = method_1_distance_to_avg(mean_diffs, None)
    pred_2 = method_2_single_layer_threshold(mean_diffs, layer_idx=5)

    method_1_preds[spk_id] = (pred_1, true_gender)
    method_2_preds[spk_id] = (pred_2, true_gender)

# Plot per-speaker correctness for each method
speakers_test_list = chosen_speakers_test
accs_m1 = [1.0 if method_1_preds[spk][0] == method_1_preds[spk][1] else 0.0
           for spk in speakers_test_list]
accs_m2 = [1.0 if method_2_preds[spk][0] == method_2_preds[spk][1] else 0.0
           for spk in speakers_test_list]

x = np.arange(len(speakers_test_list))
width = 0.35

plt.figure(figsize=(8,4))
plt.bar(x - width/2, accs_m1, width=width, label="Method 1")
plt.bar(x + width/2, accs_m2, width=width, label="Method 2")
plt.xticks(x, speakers_test_list, rotation=45)
plt.ylim(0,1.1)
plt.xlabel("Test Speaker ID")
plt.ylabel("Accuracy (0 or 1)")
plt.title("Per-Speaker Accuracy for 10 Test Speakers")
plt.legend()
plt.tight_layout()
plt.show()

print("\nDone.")
```

```python
################################################################################
# 1. INSTALL REQUIRED LIBRARIES (if not already installed)
################################################################################
# !pip install datasets
# !pip install transformers
# !pip install jiwer
# !pip install torchaudio
# !pip install scikit-learn
# !pip install matplotlib

################################################################################
# 2. IMPORTS
################################################################################
import torch
import torchaudio
import numpy as np
import random
import matplotlib.pyplot as plt

from datasets import load_dataset
from sklearn.model_selection import train_test_split
from transformers import (
    Wav2Vec2ForCTC,
    Wav2Vec2Processor,
    TrainingArguments,
    Trainer
)
from dataclasses import dataclass
from typing import Any, Dict, List, Union
from collections import defaultdict, Counter

################################################################################
# 3. LOAD THE DATASET
################################################################################
dataset = load_dataset("HamdanXI/libritts_dev_dataset_gender")["train"]
# dataset columns: ["audio", "gender", "speaker_id", "text_original", ...]

print("Full dataset size:", len(dataset))
print("Columns:", dataset.column_names)
print("Unique speaker_ids:", len(set(dataset["speaker_id"])))

################################################################################
# 4. CHOOSE 20 "TRAIN" + 10 "TEST" SPEAKERS
################################################################################
# We have male/female lists:
male_speakers_all = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "M"})
female_speakers_all = list({ex["speaker_id"] for ex in dataset if ex["gender"] == "F"})

random.seed(42)
# 1) pick 20 training: 10M + 10F
chosen_male_train = random.sample(male_speakers_all, 10)
chosen_female_train = random.sample(female_speakers_all, 10)

# Remove them from the big list so we don't re-pick them for test
for spk in chosen_male_train:
    male_speakers_all.remove(spk)
for spk in chosen_female_train:
    female_speakers_all.remove(spk)

# 2) pick 10 test: 5M + 5F
chosen_male_test = random.sample(male_speakers_all, 5)
chosen_female_test = random.sample(female_speakers_all, 5)

chosen_speakers_train = chosen_male_train + chosen_female_train
chosen_speakers_test  = chosen_male_test + chosen_female_test

print(f"Chosen training male:   {chosen_male_train}")
print(f"Chosen training female: {chosen_female_train}")
print(f"Chosen test male:       {chosen_male_test}")
print(f"Chosen test female:     {chosen_female_test}")

# Filter the dataset into two subsets: train_speakers, test_speakers
train_ds = dataset.filter(lambda x: x["speaker_id"] in chosen_speakers_train)
test_ds  = dataset.filter(lambda x: x["speaker_id"] in chosen_speakers_test)

################################################################################
# 5. LOAD PROCESSOR
################################################################################
processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")

################################################################################
# 6. PREPARE BATCH FUNCTION
################################################################################
def prepare_batch(batch):
    """
    1) Resample audio to 16k if needed.
    2) Extract raw (unpadded) input_values via processor.feature_extractor(...).
    3) Tokenize text via processor.tokenizer(...).
    """
    sr = batch["audio"]["sampling_rate"]
    audio_array = batch["audio"]["array"]
    wave_tensor = torch.tensor(audio_array, dtype=torch.float32)
    if sr != 16000:
        wave_tensor = torchaudio.functional.resample(wave_tensor, sr, 16000)
    wave_16k = wave_tensor.numpy()

    # Audio feature extraction
    audio_features = processor.feature_extractor(wave_16k, sampling_rate=16000)
    # Text tokenization
    text_tokens = processor.tokenizer(batch["text_original"])

    return {
        "speaker_id":   batch["speaker_id"],
        "gender":       batch["gender"],
        "input_values": audio_features["input_values"][0],
        "labels":       text_tokens["input_ids"]
    }

keep_cols = ["speaker_id", "gender", "input_values", "labels"]
train_ds = train_ds.map(
    prepare_batch,
    remove_columns=[col for col in train_ds.column_names if col not in keep_cols]
)

test_ds = test_ds.map(
    prepare_batch,
    remove_columns=[col for col in test_ds.column_names if col not in keep_cols]
)

################################################################################
# 7. DATA COLLATOR
################################################################################
@dataclass
class DataCollatorCTCWithPadding:
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True

    def __call__(self, features: List[Dict[str, Any]]) -> Dict[str, torch.Tensor]:
        input_values = [f["input_values"] for f in features]
        labels       = [f["labels"]       for f in features]

        audio_batch = self.processor.feature_extractor.pad(
            {"input_values": input_values},
            padding=self.padding,
            return_tensors="pt",
        )

        labels_batch = self.processor.tokenizer.pad(
            {"input_ids": labels},
            padding=self.padding,
            return_tensors="pt",
        )

        labels_tensor = labels_batch["input_ids"].masked_fill(
            labels_batch["input_ids"] == self.processor.tokenizer.pad_token_id, -100
        )

        return {
            "input_values": audio_batch["input_values"],
            "labels":       labels_tensor
        }

data_collator = DataCollatorCTCWithPadding(processor=processor, padding=True)

################################################################################
# 8. TRAIN ON SINGLE SAMPLE
################################################################################
def train_on_single_sample(
    sample,
    base_model_name="facebook/wav2vec2-base-960h",
    num_train_epochs=1,
    device="cuda" if torch.cuda.is_available() else "cpu"
):
    """
    Fine-tune on a single sample (dictionary with "input_values", "labels").
    Returns the fine-tuned model on CPU.
    """
    from datasets import Dataset

    single_ds = Dataset.from_dict({
        "input_values": [sample["input_values"]],
        "labels":       [sample["labels"]]
    })

    model = Wav2Vec2ForCTC.from_pretrained(base_model_name)
    model.to(device)
    model.freeze_feature_encoder()

    training_args = TrainingArguments(
        output_dir="temp_model",
        per_device_train_batch_size=1,
        num_train_epochs=num_train_epochs,
        evaluation_strategy="no",
        logging_steps=5,
        learning_rate=1e-5,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=single_ds,
        tokenizer=processor.feature_extractor
    )

    trainer.train()
    model.to("cpu")
    return model

################################################################################
# 9. EXTRACT LAYERWISE DIFFERENCES
################################################################################
def compute_layerwise_diff(new_model, base_model, num_layers=12):
    """
    Compare new_model vs. base_model for each Wav2Vec2 encoder layer.
    Return two lists of length num_layers: mean_diffs, std_diffs
    (Both models on CPU).
    """
    from collections import defaultdict
    layerwise_params = defaultdict(list)

    base_dict = dict(base_model.named_parameters())
    new_dict  = dict(new_model.named_parameters())

    for name, base_param in base_dict.items():
        if ("wav2vec2.encoder.layers" in name) and (name in new_dict):
            new_param = new_dict[name]
            parts = name.split(".")
            try:
                layer_idx = int(parts[3])  # e.g. "3" in "wav2vec2.encoder.layers.3..."
            except:
                continue

            if layer_idx < num_layers:
                diff = (new_param - base_param).detach().cpu().view(-1)
                layerwise_params[layer_idx].append(diff)

    mean_diffs, std_diffs = [], []
    for i in range(num_layers):
        if i in layerwise_params:
            diffs_cat = torch.cat(layerwise_params[i], dim=0)
            mean_diffs.append(diffs_cat.abs().mean().item())
            std_diffs.append(diffs_cat.abs().std().item())
        else:
            mean_diffs.append(0.0)
            std_diffs.append(0.0)

    return mean_diffs, std_diffs

################################################################################
# 10A. BUILD DIFFERENCE VECTORS FOR 20 TRAIN SPEAKERS
################################################################################
train_samples = list(train_ds)
samples_by_speaker_train = defaultdict(list)
for ex in train_samples:
    samples_by_speaker_train[ex["speaker_id"]].append(ex)

# We'll pick 1 row per speaker for the 20 training speakers
speaker_to_sample_train = {}
random.seed(1234)
for spk in chosen_speakers_train:
    spk_rows = samples_by_speaker_train[spk]
    chosen_row = random.choice(spk_rows)
    speaker_to_sample_train[spk] = chosen_row

# Load base model once
base_model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h").eval()
base_model.to("cpu")

LAYER_OFFSET = 1
NUM_LAYERS   = 12
LAYERS_TO_USE = list(range(LAYER_OFFSET, NUM_LAYERS))  # layers 1..11 (0-based indexing)

train_results = {}

print("\n=== Building difference vectors for 20 training speakers ===")
for spk_id in chosen_speakers_train:
    sample = speaker_to_sample_train[spk_id]
    print(f"Fine-tuning on speaker {spk_id} (gender={sample['gender']})")

    trained_model = train_on_single_sample(sample)
    mean_diffs, std_diffs = compute_layerwise_diff(trained_model, base_model, NUM_LAYERS)

    # Keep layers [1..11]
    mean_diffs = [mean_diffs[i] for i in LAYERS_TO_USE]
    std_diffs  = [std_diffs[i]  for i in LAYERS_TO_USE]

    train_results[spk_id] = {
        "gender": sample["gender"],
        "mean_diffs": mean_diffs,
        "std_diffs": std_diffs
    }

    del trained_model
    torch.cuda.empty_cache()

################################################################################
# 10B. AVERAGE MALE/ FEMALE DIFFERENCE VECTORS
################################################################################
male_vectors_mean  = []
male_vectors_std   = []
female_vectors_mean= []
female_vectors_std = []

for spk_id, data in train_results.items():
    if data["gender"] == "M":
        male_vectors_mean.append(data["mean_diffs"])
        male_vectors_std.append(data["std_diffs"])
    else:
        female_vectors_mean.append(data["mean_diffs"])
        female_vectors_std.append(data["std_diffs"])

# Convert to numpy for easy averaging
male_mean_avg   = np.mean(male_vectors_mean, axis=0)
male_std_avg    = np.mean(male_vectors_std,  axis=0)
female_mean_avg = np.mean(female_vectors_mean, axis=0)
female_std_avg  = np.mean(female_vectors_std,  axis=0)

################################################################################
# 11. PLOT THE TRAINING SPEAKER DIFFERENCES
################################################################################
fig1, ax1 = plt.subplots(figsize=(6,4))
fig2, ax2 = plt.subplots(figsize=(6,4))
fig3, ax3 = plt.subplots(figsize=(6,4))

layer_indices = LAYER_OFFSET + np.arange(len(LAYERS_TO_USE))  # e.g. 1..11

for spk_id, data in train_results.items():
    gender = data["gender"]
    color = "blue" if gender == "M" else "pink"

    mean_vals = data["mean_diffs"]
    std_vals  = data["std_diffs"]
    sum_vals  = [m + s for m, s in zip(mean_vals, std_vals)]

    # Plot Mean
    ax1.plot(layer_indices, mean_vals, color=color, alpha=0.7, label=gender)
    # Plot Std
    ax2.plot(layer_indices, std_vals,  color=color, alpha=0.7, label=gender)
    # Plot Mean+Std
    ax3.plot(layer_indices, sum_vals,  color=color, alpha=0.7, label=gender)

def deduplicate_legend(ax):
    handles, labels = ax.get_legend_handles_labels()
    unique_labels = []
    unique_handles = []
    for h, l in zip(handles, labels):
        if l not in unique_labels:
            unique_labels.append(l)
            unique_handles.append(h)
    ax.legend(unique_handles, unique_labels)

deduplicate_legend(ax1)
ax1.set_title("Mean Difference (Training Speakers)")
ax1.set_xlabel("Layer Index")
ax1.set_ylabel("Mean Abs Diff")

deduplicate_legend(ax2)
ax2.set_title("Std Difference (Training Speakers)")
ax2.set_xlabel("Layer Index")
ax2.set_ylabel("Std of Abs Diff")

deduplicate_legend(ax3)
ax3.set_title("Mean+Std (Training Speakers)")
ax3.set_xlabel("Layer Index")
ax3.set_ylabel("Mean+Std of Abs Diff")

plt.tight_layout()
plt.show()

################################################################################
# 12. PICK 1 ROW PER TEST SPEAKER
################################################################################
test_samples = list(test_ds)
samples_by_speaker_test = defaultdict(list)
for ex in test_samples:
    samples_by_speaker_test[ex["speaker_id"]].append(ex)

speaker_to_sample_test = {}
random.seed(1234)
for spk in chosen_speakers_test:
    spk_rows = samples_by_speaker_test[spk]
    chosen_row = random.choice(spk_rows)
    speaker_to_sample_test[spk] = chosen_row

################################################################################
# 13. DEFINE FIVE CLASSIFICATION METHODS
################################################################################
# We'll use layer 6 as mentioned (that is index=5 in our mean_diffs/std_diffs list).
LAYER_6_IDX = 5  # because we have layers [1..11] => index 0=layer1, index 5=layer6

def method_mean_euclid(new_mean):
    """
    Method 1: Compare new speaker's mean-difference vector (Euclidian distance)
              to male_mean_avg vs. female_mean_avg.
    """
    dist_male   = np.linalg.norm(new_mean - male_mean_avg)
    dist_female = np.linalg.norm(new_mean - female_mean_avg)
    return "M" if dist_male < dist_female else "F"

def method_std_euclid(new_std):
    """
    Method 2: Compare new speaker's std-difference vector (Euclidian distance)
              to male_std_avg vs. female_std_avg.
    """
    dist_male   = np.linalg.norm(new_std - male_std_avg)
    dist_female = np.linalg.norm(new_std - female_std_avg)
    return "M" if dist_male < dist_female else "F"

def method_mean_plus_std_euclid(new_mean, new_std):
    """
    Method 3: Compare new speaker's (mean+std) vector (Euclidian distance)
              to (male_mean_avg+male_std_avg) vs. (female_mean_avg+female_std_avg).
    """
    new_sum = new_mean + new_std
    male_sum = male_mean_avg + male_std_avg
    fem_sum  = female_mean_avg + female_std_avg

    dist_male   = np.linalg.norm(new_sum - male_sum)
    dist_female = np.linalg.norm(new_sum - fem_sum)
    return "M" if dist_male < dist_female else "F"

def method_single_layer_mean_threshold(new_mean, layer_idx=LAYER_6_IDX):
    """
    Method 4: Single-layer threshold (layer 6) on the mean-differences.
    Threshold = midpoint of male_mean_avg[layer_idx] and female_mean_avg[layer_idx].
    If new_mean[layer_idx] < threshold => "M", else "F".
    """
    avg_m = male_mean_avg[layer_idx]
    avg_f = female_mean_avg[layer_idx]
    threshold = 0.5 * (avg_m + avg_f)
    return "M" if new_mean[layer_idx] < threshold else "F"

def method_single_layer_std_threshold(new_std, layer_idx=LAYER_6_IDX):
    """
    Method 5: Single-layer threshold (layer 6) on the std-differences.
    Threshold = midpoint of male_std_avg[layer_idx] and female_std_avg[layer_idx].
    If new_std[layer_idx] < threshold => "M", else "F".
    """
    avg_m = male_std_avg[layer_idx]
    avg_f = female_std_avg[layer_idx]
    threshold = 0.5 * (avg_m + avg_f)
    return "M" if new_std[layer_idx] < threshold else "F"


################################################################################
# 14. EVALUATE ALL 5 METHODS ON THE 10 TEST SPEAKERS
################################################################################
method_names = [
    "mean_euclid",
    "std_euclid",
    "mean+std_euclid",
    "layer6_mean_thresh",
    "layer6_std_thresh"
]

# We'll keep track of each method's predictions as a dict: spk_id -> predicted_gender
all_preds = {m: {} for m in method_names}
true_genders = {}

print("\n=== Predicting on 10 Test Speakers with 5 Methods ===")
for spk_id in chosen_speakers_test:
    sample = speaker_to_sample_test[spk_id]
    true_gender = sample["gender"]
    true_genders[spk_id] = true_gender

    # Fine-tune single sample -> get difference vectors
    test_model = train_on_single_sample(sample)
    mean_diffs, std_diffs = compute_layerwise_diff(test_model, base_model, NUM_LAYERS)
    mean_diffs = np.array([mean_diffs[i] for i in LAYERS_TO_USE])  # length=11
    std_diffs  = np.array([std_diffs[i]  for i in LAYERS_TO_USE])  # length=11

    # Method 1: mean_euclid
    pred1 = method_mean_euclid(mean_diffs)
    all_preds["mean_euclid"][spk_id] = pred1

    # Method 2: std_euclid
    pred2 = method_std_euclid(std_diffs)
    all_preds["std_euclid"][spk_id] = pred2

    # Method 3: mean+std_euclid
    pred3 = method_mean_plus_std_euclid(mean_diffs, std_diffs)
    all_preds["mean+std_euclid"][spk_id] = pred3

    # Method 4: layer6_mean_thresh
    pred4 = method_single_layer_mean_threshold(mean_diffs, layer_idx=LAYER_6_IDX)
    all_preds["layer6_mean_thresh"][spk_id] = pred4

    # Method 5: layer6_std_thresh
    pred5 = method_single_layer_std_threshold(std_diffs, layer_idx=LAYER_6_IDX)
    all_preds["layer6_std_thresh"][spk_id] = pred5

    del test_model
    torch.cuda.empty_cache()

# Compute accuracy for each method
method_accuracies = {}
for m in method_names:
    correct = 0
    for spk_id in chosen_speakers_test:
        if all_preds[m][spk_id] == true_genders[spk_id]:
            correct += 1
    method_accuracies[m] = correct / len(chosen_speakers_test)

# Print accuracy results
for m in method_names:
    print(f"{m} accuracy: {method_accuracies[m]:.2%}")

################################################################################
# 15. PLOT PER-SPEAKER CORRECTNESS FOR EACH METHOD
################################################################################
speakers_test_list = chosen_speakers_test  # 10 test speakers
# We'll label x-axis as "spkID_gender"
speaker_labels = [f"{spk_id}_{true_genders[spk_id]}" for spk_id in speakers_test_list]

# Build array of shape [n_speakers, n_methods]
correct_matrix = np.zeros((len(speakers_test_list), len(method_names)), dtype=float)
for i, spk_id in enumerate(speakers_test_list):
    for j, m in enumerate(method_names):
        pred = all_preds[m][spk_id]
        correct_matrix[i, j] = 1.0 if (pred == true_genders[spk_id]) else 0.0

x = np.arange(len(speakers_test_list))
width = 0.15  # smaller width for 5 methods side-by-side

plt.figure(figsize=(10,4))
for j, m in enumerate(method_names):
    plt.bar(x + (j-2)*width, correct_matrix[:, j], width=width, label=m)

plt.xticks(x, speaker_labels, rotation=45)
plt.ylim(0,1.1)
plt.xlabel("Test Speaker ID (with Gender)")
plt.ylabel("Correctness (0 or 1)")
plt.title("Per-Speaker Accuracy for 10 Test Speakers (5 Methods)")
plt.legend()
plt.tight_layout()
plt.show()

print("\nDone.")
```
