```python
from datasets import Dataset, load_dataset, concatenate_datasets

dataset = load_dataset("HamdanXI/cleaned_daily_dialog_sentence")
```

```python
import re

def split_paragraph(example):
    sentences = re.split(r'(?<=[?.!])', example['dialogue'])
    return {
        'dialogue': [sentence.strip() for sentence in sentences if sentence]
    }
```

```python
nested_data = dataset['train'].map(split_paragraph, batched=False)
```

```python
flattened_datasets = [Dataset.from_dict({'dialogue': dialogues}) for dialogues in nested_data['dialogue']]
flattened_data = concatenate_datasets(flattened_datasets)
```
