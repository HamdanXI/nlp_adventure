```python
import spacy
from datasets import load_dataset

# Load the Hugging Face dataset
dataset = load_dataset("HamdanXI/cleaned_daily_dialog_sentence")
```

```python
dataset
```

```python
# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Function to extract nouns and verbs
def extract_nouns_verbs(text):
    doc = nlp(text)
    return ' '.join([token.text for token in doc if token.pos_ in ['NOUN', 'VERB']])
```

```python
# Apply the function to your dialogue column
dataset = dataset.map(lambda example: {'nouns_verbs_sentence': extract_nouns_verbs(example['dialogue'])})

# Now dataset contains the original dialogues and the new sentences with only nouns and verbs
print(dataset['train'][0])
```

```python
dataset = dataset.rename_column("dialogue", "text")
dataset = dataset.rename_column("nouns_verbs_sentence", "gloss")
```
