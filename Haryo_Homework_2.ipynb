```
def predict_something(inputs):
  predicted_output = []
  for inp in inputs:
    if inp in ['suck', 'horrible', 'stupid', 'shit']:
      predicted_output.append(2) # negative
    elif inp in ['love', 'happy', 'like']:
      predicted_output.append(0) # positive
    else:
      predicted_output.append(1) # neutral
    return predicted_output
```
