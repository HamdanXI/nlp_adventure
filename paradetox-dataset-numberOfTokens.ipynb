```python
from datasets import load_dataset

dataset = load_dataset("HamdanXI/paradetox-preprocess-maskedComments-without-INSERT-without-punctationComparision")
```

```python
dataset
```

```python
from datasets import load_dataset

def filter_rows(example):
    masked_tokens = example['masked_comment'].count('[REPLACE]') + example['masked_comment'].count('[DELETE]')
    return masked_tokens == 0

filtered_dataset_0 = dataset.filter(filter_rows)

filtered_dataset_0
```
