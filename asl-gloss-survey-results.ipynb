```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating a DataFrame from the given dataset
data = {
    "Model": ["T5-Small (ASLG-PC12)", "T5-Small (Merged Dataset)", "T5-Small (Merged - Adj & Adv)",
              "T5-Small (Daily Dialogue - ASL Glossing)", "GPT-4", "Golden Text"],
    "1st": [1, 0, 5, 1, 15, 9],
    "2nd": [7, 7, 6, 5, 3, 5],
    "3rd": [4, 4, 3, 4, 1, 3],
    "4th": [3, 5, 4, 5, 0, 2],
    "5th": [2, 5, 2, 2, 0, 1],
    "6th": [0, 0, 0, 2, 0, 0]
}

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)

# Points for each position
points = { '1st': 5, '2nd': 4, '3rd': 3, '4th': 2, '5th': 1, '6th': 0 }

# Calculate total points for each model
df['Total Points'] = df.apply(lambda row: sum(row[position] * points[position] for position in points), axis=1)

# Plotting the total points for each model
plt.figure(figsize=(12, 6))
df['Total Points'].plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel(' ')
plt.ylabel('Total Points')
plt.title('Total Points Scored by Each Model')
plt.xticks(rotation=45)

# Showing the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating the dataset
data = {
    "Model": [
        "T5-Small (ASLG-PC12)", "T5-Small (Merged Dataset)",
        "T5-Small (Merged - Adj & Adv)", "T5-Small (Daily Dialogue - ASL Glossing)",
        "GPT-4", "Golden Text"
    ],
    "Q1": ["6th", "4th", "1st", "4th", "1st", "1st"],
    "Q2": ["4th", "4th", "1st", "6th", "1st", "1st"],
    "Q3": ["2nd", "2nd", "1st", "6th", "2nd", "2nd"],
    "Q4": ["1st", "5th", "6th", "1st", "3rd", "4th"],
    "Q5": ["2nd", "4th", "4th", "6th", "1st", "3rd"],
    "Q6": ["5th", "1st", "1st", "6th", "1st", "1st"],
    "Q7": ["2nd", "5th", "4th", "6th", "1st", "3rd"],
    "Q8": ["4th", "3rd", "5th", "6th", "1st", "1st"],
    "Q9": ["3rd", "6th", "3rd", "5th", "1st", "1st"],
    "Q10": ["5th", "3rd", "3rd", "6th", "1st", "2nd"],
    "Q11": ["1st", "6th", "4th", "6th", "1st", "3rd"],
    "Q12": ["1st", "1st", "1st", "6th", "1st", "1st"],
    "Q13": ["4th", "6th", "3rd", "5th", "1st", "1st"],
    "Q14": ["3rd", "5th", "4th", "6th", "1st", "2nd"],
    "Q15": ["3rd", "3rd", "3rd", "6th", "1st", "1st"],
    "Q16": ["5th", "5th", "4th", "3rd", "2nd", "1st"],
    "Q17": ["1st", "1st", "1st", "6th", "1st", "5th"],
    "Q18": ["3rd", "3rd", "3rd", "4th", "1st", "1st"],
    "Q19": ["3rd", "5th", "5th", "4th", "1st", "1st"],
    "Q20": ["3rd", "4th", "4th", "6th", "1st", "1st"]
}

# Converting the dataset into a DataFrame
df = pd.DataFrame(data)

# Function to convert ranks to points
def rank_to_points(rank):
    rank_dict = {
        "1st": 5, "2nd": 4, "3rd": 3, "4th": 2, "5th": 1, "6th": 0
    }
    return rank_dict.get(rank, 0)

# Applying the function to convert all ranks to points
for col in df.columns[1:]:
    df[col] = df[col].apply(rank_to_points)

# Summing up the points for each model
df['Total Points'] = df.sum(axis=1)

# Plotting the bar chart
plt.figure(figsize=(10, 8))
plt.bar(df['Model'], df['Total Points'], color='skyblue')
plt.xlabel(' ')
plt.ylabel('Total Points')
plt.title('Total Points Scored by Each Model')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
